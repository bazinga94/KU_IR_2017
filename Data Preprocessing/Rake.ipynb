```python
import RAKE
import operator
new_file = open('RAKE_Fun'+'.txt','w')

sample_file = open('crawling_text'+'.txt', 'r')


for i in range (0, 1000):
    text = sample_file.readline()
    text = text.replace("-"," ")
    Rake = RAKE.Rake(RAKE.SmartStopList())  #rake로 token화
#Rake.run(text, minCharacters = 2, maxWords = 1, minFrequency = 1)
#text2 = "I hate you yeah yeah ah oh oh oh hey hey"

    list = Rake.run(text, minCharacters = 1, maxWords = 3, minFrequency = 1)  #list에 주요 문장/단어 저장
    #minCharacters = 최소문자수, maxWords = 최대 단어수, minFrequency = 최소 반복횟수

    for word in range(0,len(list)):
        for i in range(0,round(list[word][1])): #Rake 가중치 만큼 저장
            new_file.write(list[word][0]) 
            new_file.write(" ")
        print(list[word])
    new_file.write("\n")
    

new_file.close()
sample_file.close()
        
```
