```python
import nltk
from collections import Counter
import re
from nltk.corpus import wordnet

#english_vocab = set(w.lower() for w in nltk.corpus.words.words())    #영어 확인 체크
nounlist=[]

with open('./train_data/sad_train.txt','r',encoding ='utf-8') as f:
    lines = f.readlines()  #파일을 열어서 읽는다
    f.close()
    
for line in lines:
    parse = re.sub('[-=.#/!()?:$0123456789{}\[\]]', '', line)    #특수기호 제거
    
    tokens = nltk.word_tokenize(parse)     #토큰으로 만들기 
    tags = nltk.pos_tag(tokens)           #토큰 별 품사를 tagging
    
    for word, tag in tags:
        if(len(word)>2):
            if wordnet.synsets(word):   #영어인지 확인하기 위해
                
                if tag in ['NN', 'NNP', 'NNS']:     #명사를 뽑아서 nounlist에 추가 NN, NNP, NNS, NNPS, JJ, JJR, JJS
                    nounlist.append(word.lower())

counts_noun = Counter(nounlist)
```

```python
counts_noun.most_common(300)
```

```python
corpus = nltk.Text(nounlist)
corpus.dispersion_plot(['tear', 'lost', 'hate']) #단어 분포도
corpus.plot(30) #그래프
```

```python
counts_noun.most_common(400)[0][0]
A = list()
for i in range(0,400):
    A.append(counts_noun.most_common(400)[i][0]) #list에 단어만 저장
print(A)
```
